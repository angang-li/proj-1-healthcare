Visualizations: Relationship between # of Hospitals and Mortality Rate (per County)

```python
# Dependencies
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import matplotlib as mpl
```

```python
# Save path to data set in a variable
county_file = "Resources/county_data.csv"
merged_file = "Resources/final_cleaned_data.csv"
```

```python
# Use Pandas to read data
county_pd = pd.read_csv(county_file)
county_pd.head()
```

```python
# Use Pandas to read data
merged_pd = pd.read_csv(merged_file)
merged_pd.head()
```

```python
grouper = merged_pd.groupby(by=["State","County"])
grouper
```

```python
df2 = grouper["Hospital"].count().to_frame().reset_index()
df2.head()
```

```python
merged_data2 = pd.merge(county_pd, df2, on=["State", "County"], how="inner")
```

```python
merged_data2 = merged_data2.rename(columns={"Hospital" : "Number of Hospitals"})
merged_data2.head()
```

# Does # of Hospitals serving the population affect Mortality Rate?

```python
# Calculate Hospitals Per Capita and insert in merged data frame
merged_data2["Hospitals Per Capita"] = merged_data2["Population (Millions)"]/merged_data2["Number of Hospitals"]
merged_data2.head()
```

```python
# Calculate regression statistics
slope, intercept, r_value, p_value, std_err = stats.linregress(
    x=merged_data2["Hospitals Per Capita"], y=merged_data2["Crude Rate"])
```

```python
# Plot
mpl.rcParams['figure.dpi']= 100
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = ['Arial']
font_size = 13
market_size = 100
line_width = 1

sns.regplot(merged_data2["Hospitals Per Capita"], merged_data2["Crude Rate"],
           scatter_kws={"s": market_size, "facecolor": 'steelblue', 
                        "edgecolors": 'w', "linewidth": line_width},
           line_kws={"color": 'k', "linewidth": line_width})

plt.xlabel('Hospitals Per Capita', fontsize = font_size)
plt.ylabel('Crude Rate (Mortality Rate)', fontsize = font_size)
plt.tick_params(labelsize=font_size-1)
plt.grid(linestyle='--', linewidth=0.5)
plt.title("Hospitals Per Capita vs. Crude Rate (Mortality Rate)")

print(f"Linear regression: {stats.linregress(merged_data2['Hospitals Per Capita'], merged_data2['Crude Rate'])}.")
    
plt.savefig('Hospitals Per Capita_vs_Crude Rate.png', bbox_inches="tight", dpi = 300)
```

```python
# Calculate regression statistics
slope, intercept, r_value, p_value, std_err = stats.linregress(
    x=merged_data2["Hospitals Per Capita"], y=merged_data2["Age Adjusted Rate"])
```

```python
# Plot
mpl.rcParams['figure.dpi']= 100
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = ['Arial']
font_size = 13
market_size = 100
line_width = 1

sns.regplot(merged_data2["Hospitals Per Capita"], merged_data2["Age Adjusted Rate"],
           scatter_kws={"s": market_size, "facecolor": 'steelblue', 
                        "edgecolors": 'w', "linewidth": line_width},
           line_kws={"color": 'k', "linewidth": line_width})

plt.xlabel('Hospitals Per Capita', fontsize = font_size)
plt.ylabel('Crude Rate (Mortality Rate)', fontsize = font_size)
plt.tick_params(labelsize=font_size-1)
plt.grid(linestyle='--', linewidth=0.5)
plt.title("Hospitals Per Capita vs. Crude Rate (Age Adjusted Rate)")

print(f"Linear regression: {stats.linregress(merged_data2['Hospitals Per Capita'], merged_data2['Age Adjusted Rate'])}.")
    
plt.savefig('Hospitals Per Capita_vs_Age Adjusted Rate.png', bbox_inches="tight", dpi = 300)
```
