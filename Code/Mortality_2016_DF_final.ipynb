```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import patient survey data into pandas from CSV
mortality_df = pd.read_csv('Mortality_2016.csv')
mortality_df.head()
```

```python
mortality_df[["County", "State"]] = mortality_df['County'].str.split(',', expand=True)
del mortality_df['County Code']
mortality_df.head()
```

```python
clean_mortality = mortality_df[["State", "County", "Deaths", "Population", "Crude Rate",
                           "Age Adjusted Rate", "% of Total Deaths"]]
clean_mortality.head()
```

```python
clean_mortality["County"] = clean_mortality["County"].str.replace("County", "")
clean_mortality.head()
```

```python
clean_mortality["County"] = clean_mortality["County"].str.upper()
clean_mortality.head()
```

```python
# Save as csv
clean_mortality.to_csv("mortality_2016_final.csv")
```
